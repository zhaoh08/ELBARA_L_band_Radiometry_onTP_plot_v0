# Visualize seasonal variations of the Maqu ELBARA-III L-band microwave radiometry dataset during post-monsoon period


## Purpose
This notebook shows how to get access to the publised datast of the multiyear L-band microwave radiometry at the Maqu site on the eastern Tibetan Plateau in China, download the data into a folder, read data and process them and plot a figure displaying the seasonal variations of ELBARA-III TB data together with environmental variable (i.e., albedo, air temperature, soil moisture and temperature and ground temperature) data. The produced figure is similar to that in the paper "Multiyear in-situ L-band microwave radiometry of land surface processes on the Tibetan Plateau" published in a Nature Scientific Data Descriptor https://www.nature.com/articles/s41597-020-00657-1. 

The dataset is available at the figshare repository https://doi.org/10.6084/m9.figshare.12058038.v1.
The dataset is fully described in a Nature Scientific Data Descriptor https://www.nature.com/articles/s41597-020-00657-1. 

## Methodology of data measurement
#### L-band brightness temperature $T_B^p$ measurements
ELBARA-III is an L-band (1.4 GHz) Dicke-type radiometer with a dual-polarized conical horn antenna with -3 dB beamwidth of 12°. It uses a resistive load (RL), and an active cold load (ACL) as internal calibration sources to derive calibrated brightness temperature ($T_B^p$) of the ground footprints in horizontal (p=H) and vertical polarization (p=V). The absolute accuracy of ELBARA-III $T_B^p$ measurement is better than 1 K and the corresponding sensitivity is at least 0.1 K. Detailed information of $T_B^p$ derivation based on antenna measured voltage data can be found in Sci. Data paper. 

At Maqu site the ELBARA-III radiometer provides continuous measurements include angular scanning of the ground at every 30 minutes (and 15 minutes in 2016) and sky measurements. The angular range of the ground scans is between 40°-70° (relative to nadir) in steps of 5°, and sky measurements are performed at 23:55 local time every day with an observation angle of 155° (relative to nadir). The elliptic footprints at -3 dB sensitivity of the antenna are estimated according to the installation height and incidence angle.

#### Environmental variable measurements
Profile soil moisture and soil temperature are measured by 5TM sensors. 
Turbulent heat fluxes are measured by a CSAT eddy covariance system.
Meteorological data are measured by an automatic weather station (AWS) as well as air pressure and liquid precipitation measured by Geonor T-200B Series. 
Auxiliary vegetation and soil texture information were also collected during intensive field campaigns. 


## Interpretation of information from the plotted figure  
The produced figure in this notebook presents a sample of seasonal variations of the Maqu ELBARA-III L-band microwave radiometry dataset during the post-monsoon (early October to late November) period in 2016.

In the figure, three panels are plotted to display the ELBARA-III L-band brightness temperature observations in horizontal and vertical polarization at 40° incidence angle (which is consistent with the Soil Moisture Active Passive (SMAP) incidence angle), together with environmental variables (e.g., precipitation in panel 3 and other most relevant variables in panels 1 and 2) that can be used to explain the observed variations in the brightness temperature ($T_B$) in panel 3. Panel 1 displays soil moisture at 2.5 cm depth (SM_2.5 cm) and albedo, and panel 2 displays ground surface temperature (TG), air temperature at 2 m (Tair), soil temperature at 2.5 cm depth (ST_2.5 cm), and the nominal freezing point (273.15 K) as a reference to indicate whether the ground surface undergoes the freeze-thaw processes. Both panels support to appreciate and understand the variations in the ELBARA-III observations due to environmental condition change.

The figures shows that during post-monsoon period 2016, the precipitation events in the first half of October caused a reduction in ($T_B^H$, $T_B^V$) correspondingly but both gradually increased afterwards from 160 and 200 K to 180 and 220 K. This period is characterized by a gradual decrease of soil moisture from around 0.30 to 0.24 m$^3$/m$^3$ from 01/10 to 25/11 due to drying and followed by a sharp reduction to 0.11 m$^3$/m$^3$ on 26/11 due to soil freezing. TG, Tair, and ST_2.5cm all show a gradual decrease in this period, when ST_2.5cm touched the freezing point on 26/11.

## Download data into a folder on the platform
We retrieve the dataset from figshare repository and unzip it.

```python
# Define functions that read and download files
def read_public_article(id, api_url):
    request_url = f"{api_url}articles/{id}"
    print("Sending request to:", api_url)
    response = requests.get(request_url)
    status_code = response.status_code
    print(f"Response status: {status_code}")

    if status_code == 200:
        jsonResponse = response.json()
        return jsonResponse
    else:
        print("Error")
        return None

def download_file(folder, file_name, download_url):
    if os.path.exists(os.path.join(folder, file_name)):
        print(f"File {file_name} was already retrieved.")
        return
    print("Downloading ", file_name)
    response = requests.get(download_url)
    status_code = response.status_code
    if (status_code == 200):
        print(f"File {file_name} was retrieved successfully!")
        open(os.path.join(folder, file_name) , 'wb').write(response.content)
        print(f"File address: {os.path.join(folder, file_name)}")
    else:
        print(f"Could not donwload {file_name}.")
    

def download_article_files(folder, article_json):
       
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"{folder} directory was created!")

    for item in article_json['files']:
        if item['name'] == 'ELBARA-III dataset_Update.zip':     # 
            download_file(folder, item['name'], item['download_url'])      
           
article_json = read_public_article(id=id, api_url=api_url)
# The dataset is saved next to this notebook in the folder 'dataset'. you can change the path for saving the dataset in any location you would like.
folder = 'dataset'

download_article_files(folder, article_json=article_json)
#unzip the dataset
with zipfile.ZipFile(os.path.join(folder, 'ELBARA-III dataset_Update.zip'), 'r') as zip_ref:
    zip_ref.extractall(folder)
    
```

## Fill-in time period for plotting
You can change the start and end time of data collection (startT and endT) to what you desire and select them between start date and end date with the sample frequency. You may change sample frequency and get data at a 30-minutes or a hourly scale. 

```python
startT = datetime.datetime.strptime("01-10-2016 00:00", "%d-%m-%Y %H:%M")  # start date
endT = datetime.datetime.strptime("30-11-2016 23:45", "%d-%m-%Y %H:%M")    # end date
time_set = '_'+startT.strftime('%m')+startT.strftime('%d')+'_'+endT.strftime('%m')+endT.strftime('%d')+'_'+ endT.strftime('%Y')   # used to name output file
h_freq = 900       # unit: second, frequency of 15 minutes in 2016 year case, but a 30-minutes sampling frequency applies in 2017 and 2018 year cases.  
tmp = (endT-startT).total_seconds()  # the amount of seconds from start date to end date.
time_int = int(round(tmp/h_freq)+1)  # total time steps
```

# Data processing

## Read ELBARA TB file
As an example, the file of 'ELBARA_III_dataset_20160807_20170329_ELBARA_III_TB.csv' contains ELABRA-III measured $T_B^H$ and $T_B^V$ at different incidence angles. $T_B^p$ data at 40° incidence angle is extracted from the file and used in the following plot.

```python
folder = os.path.join(folder, 'ELBARA-III dataset_Update','Data')  # folder where it stores downloaded data.
filename_tb = os.path.join(folder,'ELBARA-III dataset-2016-2017ELBARA-III TB.csv') #csv file name to be read in
'''
Here the file 'ELBARA-III dataset-2016-2017ELBARA-III TB.csv' contains two times of header information. 
So the script first splits this .csv file into two files. 
'''
chunk_size = 10910  #size of rows of data to write to the csv,        
with open(filename_tb, "r") as f:
    count = 0
    lines = []
    for line in f:
        count += 1
        lines.append(line)
        if count == chunk_size:
            f1 = os.path.join(folder, 'ELBARA-III dataset-20160101-20160804ELBARA-III TB.csv')
            with open(f1, 'w') as f_out:
                f_out.writelines(lines)
                lines = []
    # write remainder
    if len(lines) > 0:
        f2 = os.path.join(folder, 'ELBARA-III dataset-20160807-20170329ELBARA-III TB.csv')
        with open(f2, 'w') as f_out:
            f_out.writelines(lines)

# Read tb file
TB_file = open(os.path.join(folder,'ELBARA-III dataset-20160807-20170329ELBARA-III TB.csv'))  # open .csv file. 
TB_F = csv.reader(TB_file, delimiter=',')  # read .csv file with delimiter of comma. 
next(TB_F)       # skip the header
TB_ref = datetime.datetime.strptime("2016-08-07 09:15", "%Y-%m-%d %H:%M") # check the original startT in .csv and the start local time in here. 
tmp2 = (startT-TB_ref).total_seconds()
T_gap = round(tmp2/h_freq)   # total time steps
print(T_gap,'data records in the read .csv file before start time')

DATE = []
TBH = []
TBV = []
i=0
 
for row in TB_F:
    if i>=T_gap and i < time_int+T_gap:  # read TB data during the period of [startT,endT]
        DATE.append(row[0])
        TBH.append(np.float64(row[2]))  # select data at 40° incidence angle
        TBV.append(np.float64(row[3]))
    i+=1

print('TB start ',DATE[0], 'TB end ',DATE[-1])
mask = numpy.asarray(numpy.isnan(TBH))  # focus on non TBH under the premise of nan value of TBV correspondingly.
mask_index = numpy.where(mask == True) 

dateList=[]
dateList_mask=[]
for x in range (0, time_int):
    dateList.append(startT+datetime.timedelta(minutes=15*x))  # change to 30*x if sampling frequency is 30 minutes. 
for i in range(0,len(mask_index[0])):
    dateList_mask.append(dateList[mask_index[0][i]])         # Period in which TBH gets non-values

df_TBH = pd.DataFrame(numpy.array(TBH),index=dateList)      # generate pandas time series data
df_TBV = pd.DataFrame(numpy.array(TBV),index=dateList)
```

## Read in situ AWS data

The file of 'ELBARA-III dataset-2016-2017MeteoData_15min_20160325_201701.csv' contains wind speed and direction, air humidity and temperature measurements at 2.5 m above ground, four radiation components (i.e., up- and downwelling shortwave and longwave radiations), as well air pressure and liquid precipitation.

```python
MeteoData_file = open(os.path.join(folder, 'ELBARA-III dataset-2016-2017MeteoData_15min_20160325_201701.csv'), encoding='utf-8-sig')

df = pd.read_csv(MeteoData_file) 
df_t = df['Datetime']

df1 = df.set_index(pd.DatetimeIndex(df_t))
Tair_insitu = df1[startT:endT]['air temperature']+273.15  # convert unit from °C to K.  

LDR = df1[startT:endT]['longwave downward']              # read longwave downward data. If you encounter error, please check the head in .csv file. 
LUR = df1[startT:endT]['longwave upwelling']
SUR = df1[startT:endT]['shortwave upwelling']
SDR = df1[startT:endT]['shortwave downward']
# RH = df1[startT:endT]['relative humidity']
pre_situ = df1[startT:endT]['Pre_situ_mm']   # 15 min - mm

# print('the length for observed shortwave downward radiation', len(SDR))
print('the first selected observed shortwave downward radiation', SDR)
```

## Derive albedo and ground temperature (TG) data based on AWS data
The albedo is calculated as the ratio of up- and downwelling shortwave radiations. High values of albedo (e.g., 0.75) indicates snowfall events. 

The surface temperature is derived from the up- and downwelling longwave radiations using Stefan–Boltzmann law. 

```python
SB = 5.67E-8  # The Stefan-Boltzmann constant
emis = 0.98   # emissivity of grassland
net_L = LUR - (1-emis)*LDR
TG_situ = (net_L/(emis*SB))**(1.0/4)          # Stefan–Boltzmann law to obtain ground temperature. 

Albedo_situ = SUR/SDR

# Simple criterias to remove noise in the albedo calcualted from longwave radiation data. You may improve it if you'd like. 
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%m/%d/%Y %H:%M')
df_night = df[df['Datetime'].dt.hour.isin(numpy.array([21,22,23,0,1,2,3,4,5]))]  # define night time
df_night.set_index('Datetime',inplace=True)  # must add inplace=True

for i in range(len(df_night.index)):   # algorithm of preprocessing derived albedo can be improved w.r.t your own case. 
    for j in range(len(SDR.index)):
        if df_night.index[i] == SDR.index[j]:
            # print df_night.index[i]
            Albedo_situ[j]=0.

for i in range(0,len(Albedo_situ)):
    if Albedo_situ[i] == numpy.inf:
        Albedo_situ[i]=0
    if Albedo_situ[i] > 1:
        Albedo_situ[i]=1.0
    if SUR[i] < 2.8:
        Albedo_situ[i]=0.
```

## Read in situ soil moisture and soil temperature (SMST) data
As an example, the file of 'ELBARA-III dataset-2016-2017SMST_LC.csv' contains soil moisture and soil temperature measured at different depths (i.e., 2 times at 2.5 cm, 5 cm, 10 cm, 20 cm, 35 cm, 60 cm, 100 cm).


```python
nlayer = 7
file_situ = open(os.path.join(folder,'ELBARA-III dataset-2016-2017SMST_LC.csv'))
header1=['Datetime','SM2_5_1','SM2_5_2','SM05','SM10','SM20','SM35','SM60','SM100','ST2_5_1','ST2_5_2','ST05','ST10','ST20','ST35','ST60','ST100']
data_situ = pd.read_csv(file_situ,sep=",")
data_situ['SM2_5'] = data_situ[['SM2_5_1', 'SM2_5_2']].mean(axis=1) # two sensors installed at different positions in the depth of 2.5 cm.  
data_situ['ST2_5'] = data_situ[['ST2_5_1', 'ST2_5_1']].mean(axis=1)
Datetime_Time = data_situ['Datetime']

data_situ1 = data_situ.set_index(pd.DatetimeIndex(Datetime_Time))
A = ['SM2_5','SM05','SM10','SM20','SM35','SM60','SM100','ST2_5','ST05','ST10','ST20','ST35','ST60','ST100']
data_situ2 = data_situ1[startT:endT][A]
SM_situ = numpy.zeros((time_int,nlayer),dtype=numpy.float32)
ST_situ = numpy.zeros((time_int,nlayer),dtype=numpy.float32)
SM_situ[:,0:nlayer] = data_situ2.iloc[:,0:nlayer]
ST_situ[:,0:nlayer] = data_situ2.iloc[:,nlayer:nlayer*2]
print('observed ST_situ ', ST_situ[0:4,:])
print('corresponding time for ST_situ ',data_situ2.index[0:4])
```

## Plot ELBARA TB data together with environmental variable (i.e., Albedo, SMST, TG, Tair) data

The plotted figure is used to analyze the consistency of the different variables, which enables further exploration and application of the data.

```python
T_s = 0  # differnt time windows can be set by changing T_s and T_e
T_e =  len(dateList)-1

fig, ax1 = plt.subplots(3,1,figsize=(10,7),dpi=150)

lns1 = ax1[0].plot(dateList[T_s:T_e],numpy.array(SM_situ)[T_s:T_e,0], 'r-',ms=6,label = 'SM_2.5cm')
ax1[0].tick_params(direction='out')
ax1[0].tick_params(top='off',bottom='on',left='on',right='off')
ax1[0].set_ylabel('SM_2.5cm (cm$^3$/cm$^3$)')

ax2 = ax1[0].twinx()
lns2 = ax2.plot(dateList[T_s:T_e],Albedo_situ[T_s:T_e],color = '0.5',linewidth=0.6,label = 'Albedo')
ax2.set_ylim(ymin = 0)
ax2.set_ylim(ymax = 1.0)
ax2.set_yticks([0,0.25,0.50,0.75,1.0])

labels = [l.get_text() for l in ax2.get_yticklabels()]
labels[0] = '0'
labels[1] = '0.25'
labels[2] = '0.50'
labels[3] = '0.75'
labels[4] = '1.0'
ax2.set_yticklabels(labels)

ax2.invert_yaxis()
ax2.set_ylabel('Albedo') 

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1[0].legend(lns, labs, loc=0,bbox_to_anchor=(0.,1.05,1,.102),ncol=2,mode="expand",borderaxespad = 0.,frameon=False)

ax1[1].plot(dateList[T_s:T_e],TG_situ[T_s:T_e], color='r',linewidth=0.6,linestyle='-',ms=6)
ax1[1].plot(dateList[T_s:T_e],Tair_insitu[T_s:T_e], color ='k',ms=6)
ax1[1].plot(dateList[T_s:T_e],ST_situ[T_s:T_e,0], color='b',linestyle='-',ms=6)
ax1[1].axhline(273.15,color = '#778899',linestyle='--')
ax1[1].legend(['TG','Tair','ST_2.5cm','273.15 K'],prop={'size': 12},loc=0,bbox_to_anchor=(0.,0.96,1,.102),ncol=4,mode="expand",borderaxespad = 0.,frameon=False)
ax1[1].set_ylabel('Temperature (K)')
ax1[1].tick_params(direction='out')
ax1[1].tick_params(top='off',bottom='on',left='on',right='off')


ax1[2].plot(dateList[T_s:T_e],numpy.array(TBH)[T_s:T_e], 'r-',ms=6)
ax1[2].plot(dateList[T_s:T_e],numpy.array(TBV)[T_s:T_e], color='orange',linestyle='-',ms=6)
# show non-value data in the plot.
line, = ax1[2].plot(dateList_mask[T_s:T_e],df_TBH.fillna(method='ffill')[mask][T_s:T_e], marker = 'o', markerfacecolor='white',linestyle='None',markersize = 4, color = 'r')  #  markevery=markers_on
line, = ax1[2].plot(dateList_mask[T_s:T_e],df_TBV.fillna(method='ffill')[mask][T_s:T_e],color='orange', marker = 'o', markerfacecolor='white',linestyle='None',markersize = 4)  # markevery=markers_on

ax1[2].legend(['$T_{B}^{H}$','$T_{B}^{V}$'],prop={'size': 12},loc=0,ncol=2,bbox_to_anchor=(0,0.97,1,.102),mode="expand",borderaxespad = 0.,frameon=False)
ax1[2].set_ylabel("$T_{B}$ (K)",fontsize=12)
ax1[2].tick_params(direction='out')
ax1[2].tick_params(top='off',bottom='on',left='on',right='off')

ax2 = ax1[2].twinx()
ax2.plot(dateList[T_s:T_e],pre_situ[T_s:T_e]*4,color = '0.5')
ax2.set_ylim(ymin = 0)
ax2.invert_yaxis()

ax2.set_ylabel('Pre (mm/h)') 

plt.gcf().autofmt_xdate()
plt.axis('tight')

# The output figure is saved next to this notebook in the folder 'Output'. you can change the path for saving the figure in any location you would like.
output_folder = 'Output'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"{output_folder} directory was created!")
Out_put=os.path.join(output_folder, 'AWS_TB_SMST_plot_2016') +time_set+ ".png"
plt.savefig(Out_put,bbox_inches='tight')
plt.show()
plt.close() 
```
